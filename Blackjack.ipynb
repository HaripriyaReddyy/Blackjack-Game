```python
############### Blackjack Project #####################

import random
#importing the random module

def deal_card():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
#In the cards list, ace is replaced with 11 and all the face cards have a value of 10
    card = random.choice(cards)
    return card
#Defining a function to randomly choose cards from the deck

def calculator_score(cards):
    if sum(cards) == 21 and len(cards) == 2:
        return 0
#if you get a blackjack in the first try, the game is over! you win
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)
#initially if the value of the cards is higher than 21, we replace the ace value from 11 to 1
    return sum(cards)
#in the above function we calculate the total score to compare the values.

def compare(user_score, computer_score):
    if user_score == computer_score:
        return "It's a draw"
    elif computer_score == 0:
        return "You lost, computer has a blackjack"
    elif user_score == 0:
        return "You win!"
    elif user_score > 21:
        return "You lose, you went higher"
    elif computer_score > 21:
        return "You win! computer went higher"
    elif computer_score > user_score:
        return "You lose"
    else:
        return "You win!"
#in the above function, multiple comparsions of the user's card value and computer's card value are done. This determines the result.

user_cards = []
computer_cards = []
#creating empty lists to store the user and computer's random cards 
is_game_over = False
for _ in range(2):
    user_cards.append(deal_card())
    computer_cards.append(deal_card())
#adding the cards to the empty list
while not is_game_over:

    user_score = calculator_score(user_cards)
    computer_score = calculator_score(computer_cards)
    print(f"Your cards {user_cards} and your score {user_score}")
    print(f"Computer's first card: {computer_cards[0]}")
#in the while loop, we are calling the calculator_score function 
    if user_score == 0 or computer_score == 0 or user_score > 21:
        is_game_over = True
#Checking if the user or the computer got a blackjack on the first try
    else:
        user_should_deal = input("Type y to get another card or n to pass:")
        if user_should_deal == "y":
            user_cards.append(deal_card())
        else:
            is_game_over = True
#the above block of code allows us to add more cards to increase the chances of winning
while computer_score != 0 and computer_score < 17:
    computer_cards.append(deal_card())
    computer_score = calculator_score(computer_cards)
#Since computer=dealer, computer's cards are hidden. The above block of code is to add cards to the computer_cards 
print(
    f"Your final hand was {user_cards} and your final score was {user_score} ")
print(
    f"Computer's final hand {computer_cards} and computers final score was {computer_score}")
print(compare(user_score, computer_score))
#The above statements are executed when the game has ended, revealing who won and by how much.
```
